Sum of a three digit number entered by user

```python
# sum of a three digit number entered by user
num = int(input("Enter a three digit number:"))
d1 = num // 100
d2 = (num // 10) % 10
d3 = num % 10

sum_digits = d1 + d2 + d3

print("Sum of digits =", sum_digits)
```

Min of three given numbers

```python
a = float(input("Enter first number"))
b = float(input("Enter second number"))
c = float(input("Enter third number"))

if a <= b and a <= c:
    min_num = a
elif b <= a and b <= c:
    min_num = b
else:
    min_num = c

print("Minimum number is:", min_num)
```

Menu Driven Program

```python
print("---- MENU ----")
print("1. Addition")
print("2. Subtraction")
print("3. Multiplication")

choice = int(input("Enter your choice: "))

if choice == 1:
    a = int(input("Enter first number: "))
    b = int(input("Enter second number: "))
    print("Sum =", a + b)

elif choice == 2:
    a = int(input("Enter first number: "))
    b = int(input("Enter second number: "))
    print("Difference =", a - b)

elif choice == 3:
    a = int(input("Enter first number: "))
    b = int(input("Enter second number: "))
    print("Product =", a * b)

else:
    print("Invalid choice")
```

Table of a given number using while loop

```python
n = int(input("Enter a number: "))

i = 1
while i <= 10:
    print(n, "x", i, "=", n * i)
    i += 1
```

Table of a given number using for loop

```python
n = int(input("Enter a number: "))

for i in range(1, 11):
    print(n, "x", i, "=", n * i)
```

Sum of all digits of a given number using while loop

```python
num = int(input("Enter a number: "))

sum_digits = 0

while num > 0:
    digit = num % 10
    sum_digits = sum_digits + digit
    num = num // 10

print("Sum of digits =", sum_digits)
```

Number Guessing Game

```python
import random

number = random.randint(1, 100)
count = 0

while True:
    guess = int(input("Guess a number: "))
    count += 1

    if guess < number:
        print("Too Low")
    elif guess > number:
        print("Too High")
    else:
        print("Correct! Attempts =", count)
        break
```
