Arithmetic Operators

```python
# Operator wo symbol hota hai jo operands (values/variables) par operation perform karta hai.
# | Operator | Meaning        | Example        |
# | -------- | -------------- | -------------- |
# | `+`      | Addition       | `10 + 5 = 15`  |
# | `-`      | Subtraction    | `10 - 5 = 5`   |
# | `*`      | Multiplication | `10 * 5 = 50`  |
# | `/`      | Division       | `10 / 5 = 2.0` |
# | `%`      | Modulus        | `10 % 3 = 1`   |
# | `//`     | Floor division | `10 // 3 = 3`  |
# | `**`     | Power          | `2 ** 3 = 8`   |
print(10 + 3)
print(10 // 3)
print(2 ** 4)
```

Relational / Comparison Operators

```python
# Result True ya False hota hai
# | Operator | Meaning          |
# | -------- | ---------------- |
# | `==`     | Equal            |
# | `!=`     | Not equal        |
# | `>`      | Greater          |
# | `<`      | Less             |
# | `>=`     | Greater or equal |
# | `<=`     | Less or equal    |
print(10 > 5)    # True
print(10 == 5)   # False
```

Logical Operators

```python
# | Operator | Meaning      |
# | -------- | ------------ |
# | `and`    | Both true    |
# | `or`     | Any one true |
# | `not`    | Reverse      |
x = 10
print(x > 5 and x < 20)   # True
print(not(x > 5))        # False
```

Assignment Operators

```python
# | Operator | Example  |
# | -------- | -------- |
# | `=`      | `x = 5`  |
# | `+=`     | `x += 2` |
# | `-=`     | `x -= 2` |
# | `*=`     | `x *= 2` |
# | `/=`     | `x /= 2` |
# | `%=`     | `x %= 2` |
x = 10
x += 5
print(x)   # 15
```

Bitwise Operators

```python
# | Operator | Meaning     |    |
# | -------- | ----------- | -- |
# | `&`      | AND         |    |
# | `        | `           | OR |
# | `^`      | XOR         |    |
# | `~`      | NOT         |    |
# | `<<`     | Left shift  |    |
# | `>>`     | Right shift |    |
a = 5   # 101
b = 3   # 011
print(a & b)   # 1
```

Membership Operators

```python
# | Operator | Meaning     |
# | -------- | ----------- |
# | `in`     | present     |
# | `not in` | not present |
lst = [1,2,3]
print(2 in lst)      # True
print(5 not in lst)  # True
```

Identity Operators

```python
# | Operator | Meaning          |
# | -------- | ---------------- |
# | `is`     | same object      |
# | `is not` | different object |
a = [1,2]
b = a
c = [1,2]

print(a is b)   # True
print(a is c)   # False
```

Simple if Statement

```python
# if condition:
#     statement
x = 10
if x > 0:
    print("Positive number")
```

if–else Statement

```python
# if condition:
#     statement
# else:
#     statement
num = int(input("Enter a number: "))

if num % 2 == 0:
    print("Even number")
else:
    print("Odd number")
```

if–elif–else (Multiple Conditions)

```python
# if condition1:
#     statement
# elif condition2:
#     statement
# else:
#     statement
marks = int(input("Enter marks: "))

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 50:
    print("Grade C")
else:
    print("Fail")
```

Nested if–else (if ke andar if)

```python
num = int(input("Enter a number: "))

if num > 0:
    if num % 2 == 0:
        print("Positive Even")
    else:
        print("Positive Odd")
else:
    print("Negative number")
```

Short Hand if–else (Ternary Operator)

```python
a = 10
b = 20

max = a if a > b else b
print(max)
```

while Loop (CONDITION BASED)

```python
# while condition:
#     statement

i = 1
while i <= 5:
    print(i)
    i += 1
```

for Loop (SEQUENCE BASED)

```python
# for variable in sequence:
#     statement

for i in range(1, 6):
    print(i)
```

range()

```python
# range(start, stop, step)

range(5)        # 0 to 4
range(1, 6)     # 1 to 5
range(1, 10, 2) # 1 3 5 7 9
```

break Statement

```python
for i in range(1, 10):
    if i == 5:
        break
    print(i)
```

continue Statement

```python
for i in range(1, 6):
    if i == 3:
        continue
    print(i)
```

pass Statement

```python
for i in range(3):
    pass
```

Nested Loop (Loop ke andar loop)

```python
for i in range(1, 4):
    for j in range(1, 3):
        print(i, j)
```
