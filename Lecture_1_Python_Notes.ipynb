Python Output

```python
#Program ka result screen par dikhana
#python is a case sensitive language.
#Python me output ke liye mainly print() use hota hai.

print("Hello World")#normal print

print(7)#number k liye print
print(7,7)#ek se zyada nos.
print(True)#boolean values
print("hello",1,4.5,True)#more than one mixed values

print("hello",1,4.5,True,sep="/")#using sep
print("hello",end="-")#using end
print("world")

print("Hello\nWorld")#for new line
print("Name:\tShagun")#for tab
print("Hello\"World")#for double quote
print("Hello\\World")#for backslash
```

Data Types

```python
# Data type batata hai ki variable ke andar kis type ka data store hai
# Python me data type automatically detect hota hai.

print(7)#int
print(8.5)#decimal
print(True)#boolean
print("hello world")#string
print(5+6j)#complex
print([1,2,3,4,5])#list
print((1,2,3,4,5))#tuple
print({1,2,3,4,5})#sets
print({'name':'shagun','age':20})#dictionary

type([1,2,3])
```

Variables

```python
#Variable ek naam hota hai jo memory me kisi value ko refer karta hai.
#Python me variable banane ke liye data type likhne ki zarurat nahi hoti.
# Letter (a–z, A–Z) se start
# _ se start ho sakta hai
# Number se start nahi ho sakta
# Space allowed nahi
# Keywords allowed nahi

name = "Shagun"
print(name)

a=5
b=6
print(a+b)

a,b,c = 1,2,3
print(a,b,c)

a=b=c=5
print(a,b,c)

x = 10
print(type(x))

x = 10
print(id(x))

x = 5
x = x + 2
print(x)

x = 10
del x

# Dynamic typing - jisme hum data type nhi btate hai . python ka interpreter automatically smjh jata hai .
# Static typing - jisme hum data type btate hai . python ka interpreter automatically nhi smjhta .
# dynamic binding - python ka variable ek hi program me kisi bhi type ka data type hold kr skta h. at run time
# static binding - at compile time
```

Type Conversion

```python
# Ek data type ko doosre data type me badalna hi type conversion kehlata hai.

# Implicit - Python automatically ek data type ko doosre me convert karta hai (jab data loss ka risk nahi hota).
a = 10      # int
b = 2.5     # float
c = a + b
print(c)
print(type(c))

#Explicit - Jab programmer khud manually type convert karta hai.
int("20")     # 20
```

Literals

```python
# Literal wo fixed value hoti hai jo directly program me likhi jaati hai.
# Literal ki value change nahi hoti (jab tak hum code me change na karein).
x = 0b1010   # Binary
y = 0o12     # Octal
z = 0xA      # Hexadecimal
```
