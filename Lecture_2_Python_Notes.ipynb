Arithmetic Operators

```python
# Operator wo symbol hota hai jo operands (values/variables) par operation perform karta hai.
# | Operator | Meaning        | Example        |
# | -------- | -------------- | -------------- |
# | `+`      | Addition       | `10 + 5 = 15`  |
# | `-`      | Subtraction    | `10 - 5 = 5`   |
# | `*`      | Multiplication | `10 * 5 = 50`  |
# | `/`      | Division       | `10 / 5 = 2.0` |
# | `%`      | Modulus        | `10 % 3 = 1`   |
# | `//`     | Floor division | `10 // 3 = 3`  |
# | `**`     | Power          | `2 ** 3 = 8`   |
print(10 + 3)
print(10 // 3)
print(2 ** 4)
```

Relational / Comparison Operators

```python
# Result True ya False hota hai
# | Operator | Meaning          |
# | -------- | ---------------- |
# | `==`     | Equal            |
# | `!=`     | Not equal        |
# | `>`      | Greater          |
# | `<`      | Less             |
# | `>=`     | Greater or equal |
# | `<=`     | Less or equal    |
print(10 > 5)    # True
print(10 == 5)   # False
```

Logical Operators

```python
# | Operator | Meaning      |
# | -------- | ------------ |
# | `and`    | Both true    |
# | `or`     | Any one true |
# | `not`    | Reverse      |
x = 10
print(x > 5 and x < 20)   # True
print(not(x > 5))        # False
```

Assignment Operators

```python
# | Operator | Example  |
# | -------- | -------- |
# | `=`      | `x = 5`  |
# | `+=`     | `x += 2` |
# | `-=`     | `x -= 2` |
# | `*=`     | `x *= 2` |
# | `/=`     | `x /= 2` |
# | `%=`     | `x %= 2` |
x = 10
x += 5
print(x)   # 15
```

Bitwise Operators

```python
# | Operator | Meaning     |    |
# | -------- | ----------- | -- |
# | `&`      | AND         |    |
# | `        | `           | OR |
# | `^`      | XOR         |    |
# | `~`      | NOT         |    |
# | `<<`     | Left shift  |    |
# | `>>`     | Right shift |    |
a = 5   # 101
b = 3   # 011
print(a & b)   # 1
```

Membership Operators

```python
# | Operator | Meaning     |
# | -------- | ----------- |
# | `in`     | present     |
# | `not in` | not present |
lst = [1,2,3]
print(2 in lst)      # True
print(5 not in lst)  # True
```

Identity Operators

```python
# | Operator | Meaning          |
# | -------- | ---------------- |
# | `is`     | same object      |
# | `is not` | different object |
a = [1,2]
b = a
c = [1,2]

print(a is b)   # True
print(a is c)   # False
```

Simple if Statement

```python
# if condition:
#     statement
x = 10
if x > 0:
    print("Positive number")
```

if–else Statement

```python
# if condition:
#     statement
# else:
#     statement
num = int(input("Enter a number: "))

if num % 2 == 0:
    print("Even number")
else:
    print("Odd number")
```

if–elif–else (Multiple Conditions)

```python
# if condition1:
#     statement
# elif condition2:
#     statement
# else:
#     statement
marks = int(input("Enter marks: "))

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 50:
    print("Grade C")
else:
    print("Fail")
```

Nested if–else (if ke andar if)

```python
num = int(input("Enter a number: "))

if num > 0:
    if num % 2 == 0:
        print("Positive Even")
    else:
        print("Positive Odd")
else:
    print("Negative number")
```

Short Hand if–else (Ternary Operator)

```python
a = 10
b = 20

max = a if a > b else b
print(max)
```

while Loop (CONDITION BASED)

```python
# while condition:
#     statement

i = 1
while i <= 5:
    print(i)
    i += 1
```

for Loop (SEQUENCE BASED)

```python
# for variable in sequence:
#     statement

for i in range(1, 6):
    print(i)
```

```python
for i in range(1, 4):
    for j in range(1, 3):
        print(i, j)
```

Sum of a three digit number entered by user

```python
# sum of a three digit number entered by user
num = int(input("Enter a three digit number:"))
d1 = num // 100
d2 = (num // 10) % 10
d3 = num % 10

sum_digits = d1 + d2 + d3

print("Sum of digits =", sum_digits)
```

Min of three given numbers

```python
a = float(input("Enter first number"))
b = float(input("Enter second number"))
c = float(input("Enter third number"))

if a <= b and a <= c:
    min_num = a
elif b <= a and b <= c:
    min_num = b
else:
    min_num = c

print("Minimum number is:", min_num)
```

Menu Driven Program

```python
print("---- MENU ----")
print("1. Addition")
print("2. Subtraction")
print("3. Multiplication")

choice = int(input("Enter your choice: "))

if choice == 1:
    a = int(input("Enter first number: "))
    b = int(input("Enter second number: "))
    print("Sum =", a + b)

elif choice == 2:
    a = int(input("Enter first number: "))
    b = int(input("Enter second number: "))
    print("Difference =", a - b)

elif choice == 3:
    a = int(input("Enter first number: "))
    b = int(input("Enter second number: "))
    print("Product =", a * b)

else:
    print("Invalid choice")
```

Table of a given number using while loop

```python
n = int(input("Enter a number: "))

i = 1
while i <= 10:
    print(n, "x", i, "=", n * i)
    i += 1
```

Table of a given number using for loop

```python
n = int(input("Enter a number: "))

for i in range(1, 11):
    print(n, "x", i, "=", n * i)
```

Sum of all digits of a given number using while loop

```python
num = int(input("Enter a number: "))

sum_digits = 0

while num > 0:
    digit = num % 10
    sum_digits = sum_digits + digit
    num = num // 10

print("Sum of digits =", sum_digits)
```

Number Guessing Game

```python
import random

number = random.randint(1, 100)
count = 0

while True:
    guess = int(input("Guess a number: "))
    count += 1

    if guess < number:
        print("Too Low")
    elif guess > number:
        print("Too High")
    else:
        print("Correct! Attempts =", count)
        break
```
