String Indexing and Slicing

```python
# String ke har character ko ek number (index) diya jata hai, jisse hum us character ko access kar sakte hain.
# Index 0 se start hota hai

#Basic Indexing (Positive Index)
s = "Python"
print(s[0])   # P
print(s[3])   # h

# Negative Indexing
# End se start hoti hai -1
print(s[-1])   # n
print(s[-2])   # o

# String Slicing (Index ka Range)
# string[start : end]
print(s[0:4])   # Pyth
print(s[2:5])   # tho
print(s[:3])    # Pyt
print(s[3:])    # hon

# Slicing with Step
# string[start : end : step]
print(s[::2])    # Pto
print(s[1::2])   # yhn

# Reverse String Using Indexing
print(s[::-1])
print(s[len(s)-1])
print(s[-1])
```

Editing and Deleting Strings

```python
# Python strings immutable hoti hain
# String ka koi character directly change (edit) nahi kar sakte
# Delete bhi direct character level par nahi hota

# Replace a character
s = "Hello"
s = 'h' + s[1:]
print(s)

# Replace using replace()
s = "I love Java"
s = s.replace("Java", "Python")
print(s)

# Editing String Using Slicing
s = "Python"
new_s = s[:2] + "zz" + s[4:]
print(new_s)

# Delete entire string using del
s = "Python"
del s

# Remove characters using slicing
s = "Python"
s = s[1:]
print(s)

# Remove substring using replace()
s = "Hello World"
s = s.replace("World", "")
print(s)

# Deleting Characters Using Loop
s = "banana"
new_s = ""

for ch in s:
    if ch != 'a':
        new_s += ch

print(new_s)
```

Operations on Strings

```python
# String operations wo actions hain jo hum strings par perform karte hain,jaise jodna, repeat karna, compare karna, search karna, etc.

s1 = "Hello"
s2 = "World"
print(s1 + " " + s2)

print("Hi " * 3)

print("apple" == "apple")   # True
print("apple" < "banana")  # True
print("Apple" < "apple")   # True

print("Hello" and "World")   # World
print("" and "Python")       # ""
print("Hello" or "")         # Hello
print(not "Python")          # False
print(not "")                # True

s = "Python"
for ch in s:
    print(ch)

for i in range(len(s)):
    print(s[i])

s = "Python"
count = 0
for ch in s:
    if ch in "aeiouAEIOU":
        count += 1
print("Vowels =", count)

print("Py" in "Python")        # True
print("Java" in "Python")      # False
print("Java" not in "Python")  # True
```

String Functions

```python
# String functions (methods) wo built-in methods hote hain jo string objects par directly apply hote hain aur nayi string return karte hain (kyunki strings immutable hoti hain).

s = "Python"
print(min(s))

s = "Python"
print(max(s))

s = "python"
print(sorted(s))

s = "python"
new_s = "".join(sorted(s))
print(new_s)

s = "python programming"
print(s.capitalize())
print(s.title())
print(s.upper())
print(s.lower())
s2 = "PyThOn"
print(s2.swapcase())

print(s.count('p'))
print(s.count("pro"))
print(s.find("pro"))
print(s.find("java"))
print(s.index("pro"))

s = "python_programming.py"
print(s.startswith("python"))
print(s.startswith("java"))
print(s.endswith(".py"))
print(s.endswith(".java"))

name = "Shagun"
age = 20
print("My name is {} and age is {}".format(name, age))

print("I love {0} and {1}".format("Python", "Java"))
print("I love {1} and {0}".format("Python", "Java"))

print("Name: {n}, Age: {a}".format(n="Shagun", a=20))

print("Python123".isalnum())   # True
print("Python 123".isalnum()) # False
print("Python@123".isalnum()) # False

print("Python".isalpha())     # True
print("Python123".isalpha()) # False
print("Hello World".isalpha()) # False

print("123".isdigit())     # True
print("12.3".isdigit())   # False
print("123a".isdigit())   # False

print("age".isidentifier())     # True
print("_name".isidentifier())  # True
print("2age".isidentifier())   # False
print("for".isidentifier())    # True (identifier, but keyword)

s = "Python is easy"
print(s.split())

s = "one two three four"
print(s.split(" ", 2))

words = ['Python', 'is', 'easy']
print(" ".join(words))

s = "I love Java"
print(s.replace("Java", "Python"))

s = "banana"
print(s.replace("a", "o", 2))

s = "   hello world   "
print(s.strip())

s = "***hello***"
print(s.strip("*"))

print(s.lstrip())

print(s.rstrip())
```
