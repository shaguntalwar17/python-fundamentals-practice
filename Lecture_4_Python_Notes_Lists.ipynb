arrays vs lists

```python
# List → Python ka built-in data type
# Ordered collection
# Different data types allowed
# Mutable
# Built-in
# Heterogeneous (mixed data types)
# Dynamic size
# Easy to use
# can have duplicates
# can be nested

# Array → Python me module se aata hai (array module / NumPy array)
# Same type ke elements ka collection
# Homogeneous (single data type)
# Module import karna padta hai
# Same data type only
# More memory efficient
# Faster numeric operations
# Slightly complex syntax

# Use LIST when:
# Data types mixed ho
# Normal programming / beginners
# Flexibility chahiye

# Use ARRAY when:
# Same type ka numeric data
# Memory & speed important ho
# Scientific / numeric calculations

# List → references store karti hai
# Array → actual values store karta hai (continuous memory)
```

Creating Lists

```python
lst = [1, 2, 3, 4]
print(lst)

lst = []
print(lst)

lst = list((1, 2, 3))
print(lst)

lst = list(range(1, 6))
print(lst)

lst = [10, "Python", 3.5, True]
print(lst)

lst = list("Python")
print(lst)

matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

cube = [
    [
        [1, 2],
        [3, 4]
    ],
    [
        [5, 6],
        [7, 8]
    ]
]
```

Accessing lists

```python
lst = [10, 20, 30, 40, 50]
print(lst[0])   # 10
print(lst[3])   # 40

print(lst[-1])   # 50
print(lst[-2])   # 40

print(lst[1:4])   # [20, 30, 40]
print(lst[:3])    # [10, 20, 30]
print(lst[2:])    # [30, 40, 50]

print(lst[::2])   # [10, 30, 50]
print(lst[1::2])  # [20, 40]

print(lst[::-1])

matrix = [
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
]
print(matrix[0][0])   # 10
print(matrix[1][2])   # 60
print(matrix[2][1])   # 80

print(matrix[0:2])
print(matrix[1][0:2])

cube = [
    [           # Layer 0
        [1, 2],
        [3, 4]
    ],
    [           # Layer 1
        [5, 6],
        [7, 8]
    ]
]
print(cube[0][0][0])   # 1
print(cube[0][1][1])   # 4
print(cube[1][0][1])   # 6
print(cube[1][1][0])   # 7

print(cube[0:1])

print(cube[1][0:1])

print(cube[1][1][0:1])
```

Adding items in List

```python
lst = [10, 20, 30]
lst.append(40)
print(lst)

lst = [10, 20, 30]
lst.insert(1, 15)
print(lst)

lst = [1, 2]
lst.extend([3, 4, 5])
print(lst)

lst = [10, 20, 30]
lst[1:1] = [15, 16]
print(lst)
```

editing items in list

```python
lst = [10, 20, 30, 40]
lst[1] = 99
print(lst)

lst = [10, 20, 30, 40, 50]
lst[1:4] = [99, 88]
print(lst)

lst = [1, 2, 3]
lst[:] = [10, 20, 30]
print(lst)

matrix = [
    [1, 2, 3],
    [4, 5, 6]
]

matrix[0][1] = 99
print(matrix)
```

deleting lists

```python
lst = [10, 20, 30, 20]
lst.remove(20)
print(lst)

lst = [10, 20, 30]
lst.pop(1)
print(lst)

lst = [10, 20, 30]
del lst[1]
print(lst)

lst = [10, 20, 30, 40, 50]
del lst[1:4]
print(lst)

lst = [1, 2, 3]
lst.clear()
print(lst)
```

 Operators on Lists

```python
a = [1, 2]
b = [3, 4]
print(a + b)

lst = [1, 2]
print(lst * 3)

print([1,2] == [1,2])     # True
print([1,2] < [1,3])     # True
print([1,5] > [1,3])     # True

print([1,2] and [3,4])   # [3,4]
print([] and [1,2])     # []
print([1,2] or [])      # [1,2]
print(not [])           # True
print(not [5])          # False

lst = [10, 20, 30]
print(20 in lst)        # True
print(40 not in lst)   # True

a = [1,2]
b = a
c = [1,2]
print(a is b)      # True
print(a is c)      # False
print(a == c)     # True

lst = [10, 20, 30, 40]
for x in lst:
    print(x)

lst = [10, 20, 30]
for i in range(len(lst)):
    print(i, lst[i])

lst = ["a", "b", "c"]
for i, val in enumerate(lst):
    print(i, val)

lst = [5, 10, 15]
i = 0
while i < len(lst):
    print(lst[i])
    i += 1

matrix = [
    [1, 2, 3],
    [4, 5, 6]
]

for row in matrix:
    for item in row:
        print(item, end=" ")
    print()

lst = [1, 2, 3, 4]
for i in range(len(lst)):
    lst[i] = lst[i] * 2
print(lst)

lst = [10, 20, 30, 40]
for x in lst[:]:
    if x > 20:
        lst.remove(x)
print(lst)
```

List Function: len(), min(), max(), sorted(), count(), index(), reverse(), sort(), copy()

```python
lst = [10, 20, 30, 40]
print(len(lst))

print(min(lst))

print(max(lst))

print(sorted(lst))
print(lst)

print(lst.count(20))

print(lst.index(30))

lst.reverse()
print(lst)

lst.sort()
print(lst)

new_lst = lst.copy()
print(new_lst)
```

List Comprehension

```python
# List comprehension ek short aur powerful way hai list banane ka, jo for loop aur if condition ko ek hi line me likhne deta hai.
squares = [x*x for x in range(1, 6)]
print(squares)

even = [x for x in range(1, 11) if x % 2 == 0]
print(even)

lst = ["Even" if x % 2 == 0 else "Odd" for x in range(1, 6)]
print(lst)

nums = [1, 2, 3, 4]
double = [x*2 for x in nums]
print(double)

s = "Python"
chars = [ch for ch in s]
print(chars)

matrix = [[1, 2], [3, 4]]
flat = [item for row in matrix for item in row]
print(flat)
```

itemwise and indexwise traversal of lists

```python
lst = [10, 20, 30, 40]
for item in lst:
    print(item)

lst = [10, 20, 30, 40]
for i in range(len(lst)):
    print(i, lst[i])
```

zip function

```python
# zip() function multiple iterables (list, tuple, etc.) ke elements ko pair karta hai aur tuple ke form me combine karta hai.
a = [1, 2, 3]
b = ['a', 'b', 'c']
z = zip(a, b)
print(list(z))

names = ["A", "B", "C"]
marks = [80, 90, 85]
for n, m in zip(names, marks):
    print(n, m)

pairs = [(1, 'a'), (2, 'b'), (3, 'c')]
x, y = zip(*pairs)
print(list(x))
print(list(y))
```
